```python
## 周k选股
## 条件：
# 1.上周周k长上影
# 2.本周价位处于(close+low)/2时形成买点
```

```python
## 获取数据源
## 从导出的表格中获取同花顺最近半年涨停超过N次的非ST股票数据
## 条件：非st股票 半年内涨停超过3次
import xlrd
data = xlrd.open_workbook('/data/stock_new/half_year_good_stocks_3.xlsx')
table = data.sheets()[0]
code_list = table.col_values(0)[1::]
codes = [code.split('.')[0] for code in code_list]
print(len(codes))
print(codes)
```

```python
import requests
import json

def isLongUpShadow(close, high):
    return (high - close) / close > 0.05
    
def getSuffixCode(code):
    if code[0] == '6':
        return code + '1'
    else: 
        return code + '2'
def isTargetStock(code_id):
    easy_url = 'http://pdfm.eastmoney.com/EM_UBG_PDTI_Fast/api/js?rtntype=5&token=4f1862fc3b5e77c150a2b985b12db0fd&type=wk&authorityType=fa&id='
    final_url = easy_url + getSuffixCode(code_id)
    response = requests.get(final_url)
    data_str = response.text.split('(')[-1].split(')')[0]
    data = json.loads(data_str)['data']
    data_array = data[-1].split(',')
    close_p = float(data_array[2])
    open_p = float(data_array[1])
    high_p = float(data_array[3])
    low_p = float(data_array[4])
    return isLongUpShadow(close_p, high_p)
print('done')
```

```python
## 计算上周长上影股票
longCodes = []
failed_codes = []
for code in codes:
    try:
        if isTargetStock(code):
            longCodes.append(code)
    except BaseException as e:
        failed_codes.append(code)
        pass
    continue
```

```python
print(longCodes)
```

```python
codes_str = ' '.join(longCodes)
with open("/data/stock_new/weeklongshadow.txt","w") as f:
    f.writelines(codes_str)
```
