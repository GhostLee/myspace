```python
## 获取财务数据
import tushare as ts
import numpy as np
from pandas import DataFrame
import time
from pyecharts import Bar, Kline, Line, Grid


try:
    settingFile = open('/data/stock-setting.txt', 'r')
    settings = settingFile.read()
    token = settings.split()[0]
    secretUrl = settings.split()[1]
    settingFile.close()
except IOError:
    print('File is not accessible.')
pro = ts.pro_api(token)
 
df = pro.stock_basic(list_status = 'L')
ts_codes = df.ts_code.values
excelColumnNum = 5
result = []
failedCodes = []
for index, row in df.iterrows():
    if index % 200 == 0:
        print(str(index) + '个股票已经被处理')
    try:
        code = row.ts_code
        row_result = [0] * excelColumnNum
        df = pro.query('fina_indicator', ts_code=code, start_date='20181201', fields = 'ts_code, ann_date, end_date, netprofit_yoy')
        year_netprofit_yoy = df[df.end_date == '20181231'].iloc[0, 3]
        first_quarter_netprofit_yoy = df[df.end_date == '20190331'].iloc[0, 3]
        nameHistory = pro.namechange(ts_code = row.ts_code, fields='name')
        name = nameHistory.iloc[0, 0]

        row_result[0] = index
        row_result[1] = code
        row_result[2] = name
        row_result[3] = str(round(float(year_netprofit_yoy), 1)) + '%'
        row_result[4] = str(round(float(first_quarter_netprofit_yoy), 1)) + '%'
        result.append(row_result)
        time.sleep(0.5)
    except BaseException as e:
        failedCodes.append(code)
        pass
    continue
    
print('success')
print(result)
    
```

```python
final_result = []
for i in range(0, len(result)):
    target = result[i]
    a = target[3].split('%')[0]
    b = target[4].split('%')[0]
    target[3] = a
    target[4] = b
    final_result.append(target)
print('success')
```

```python
directory = '/data/stock/'
fileName = 'stock_profit_1.xls'
columnName = '编号\t股票代码\t股票名称\t年净利润同比\t一季度净利润同比\n'
output = open(directory + fileName,'w',encoding='gbk')
output.write(columnName)
for i in range(len(final_result)):
    for j in range(len(final_result[i])):
        output.write(str(final_result[i][j]))
        output.write('\t')  
    output.write('\n')       
output.close()
```
