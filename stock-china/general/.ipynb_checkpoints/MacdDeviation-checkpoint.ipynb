```python
# import tushare as ts
import numpy as np
import pandas as pd
from pandas import DataFrame
import time
import baostock as bs

# # baostock获取前复权收盘价 
# lg = bs.login()
# rs = bs.query_history_k_data_plus("sz.300542", "date, code, close",  adjustflag="2")
# data_list = []
# while (rs.error_code == '0') & rs.next():
#     data_list.append(rs.get_row_data())
# df = pd.DataFrame(data_list, columns=rs.fields)
# print(df)


try:
    settingFile = open('/data/stock-setting.txt', 'r')
    settings = settingFile.read()
    token = settings.split()[0]
    secretUrl = settings.split()[1]
    settingFile.close()
except IOError:
    print('File is not accessible.')
pro = ts.pro_api(token)

SHORT = 10
LONG = 20
MID = 5
s1 = (SHORT-1)/(SHORT+1) 
s2 = 2/(SHORT+1) 
l1 = (LONG-1)/(LONG+1)
l2 = 2/(LONG+1)
m1 = (MID-1)/(MID+1)
m2 = 2/(MID+1)

def CalcMacd(ema_s, ema_l, dea, close):
    emas_s = ema_s * s1 + close * s2
    emas_l = ema_l * l1 + close * l2
    diff = emas_s - emas_l
    dea = dea * m1 + diff * m2
    bar = 2 * (diff - dea)
    return [emas_s, emas_l, diff, dea, bar]

#ts_code = '300542.SZ'  ##新晨科技
ts_code = '000830.SZ'
df = ts.pro_bar(ts_code=ts_code, adj='qfq', pro_api = pro)

data = df.iloc[::-1]
result = DataFrame([], columns=['ts_code', 'trade_time', 'close', 'macd', 'macd_diff', 'dea', 'ema_s', 'ema_l'])
ema_s = 0 
ema_l = 0
dea = 0
i = 0
for index, row in data.iterrows(): 
    close = row.close
    if i == 0:
        ema_s = close 
        ema_l = close
        dea = 0
        i = 1
        continue
    dailyMacd = CalcMacd(ema_s, ema_l, dea, close)
    ema_s = dailyMacd[0] 
    ema_l = dailyMacd[1]
    dea = dailyMacd[3]
    result = result.append({'ts_code':ts_code, 'trade_time':row.trade_date, 'close':row.close, 'macd':dailyMacd[4], 'macd_diff':dailyMacd[2], 'dea':dea, 'ema_s':ema_s, 'ema_l':ema_l}
                        , ignore_index = True)
    i = i + 1
print('success')
```
