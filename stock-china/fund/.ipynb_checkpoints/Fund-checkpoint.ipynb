```python
#### 获取基金基础信息 ####
import tushare as ts
import time
import pandas as pd

try:
    settingFile = open('/data/stock-setting.txt', 'r')
    settings = settingFile.read()
    token = settings.split()[0]
    secretUrl = settings.split()[1]
    settingFile.close()
except IOError:
    print('File is not accessible.')
pro = ts.pro_api(token)
pd.set_option('display.unicode.ambiguous_as_wide', True)
pd.set_option('display.unicode.east_asian_width', True)
pd.set_option('display.width', 5000)
fund = pro.fund_basic(market = 'O', fields = 'ts_code, name, fund_type, issue_date, due_date, delist_date, min_amount, issue_amount, m_fee, c_fee, benchmark')
print('success')
```

```python
#### 筛选基金 ####
from pandas import DataFrame 
import traceback

print('共有场外基金共', len(fund), '个')

##筛选成立超过5年的基金
matureFund = fund[(fund.issue_date <= '20140401') 
                  & ((fund.due_date.isna()) |(fund.due_date >= '20190402'))
                  & ((fund.delist_date.isna()) |(fund.delist_date >= '20190402'))]
print('成立超过5年的基金共', len(matureFund), '个')

##筛选资金量
bigFund = matureFund[matureFund.issue_amount >= 1]
print('发行份额大于20亿的基金共', len(bigFund), '个')

##获取历史净值数据
funds = DataFrame([], columns=['ts_code', 'startTime', 'current', 'oneMonth', 'threeMonth', 'halfYear', 'oneYear', 'twoYear'])
unusedCount = 0;
try:
    for index, row in bigFund.iterrows():
        tsCode = row.ts_code
        nav = pd.read_csv('/data/stock/data/fund/history/' + tsCode + '.csv', index_col = 0) 
        if nav.shape[0] <= 500:
            unusedCount = unusedCount + 1 
            continue
            
        currentNav = nav.iloc[0, 3]
        oneMonthNav = nav.iloc[20, 3]
        threeMonthNav = nav.iloc[55, 3]
        halfYearNav = nav.iloc[125, 3]
        oneYearNav = nav.iloc[250, 3]
        twoYearNav = nav.iloc[500, 3]
        bottomNav = nav.iloc[773, 3]
        topNav = nav.iloc[926, 3]
        
        startTime = nav.iloc[926, 1] 

        funds = funds.append({'ts_code':tsCode, 'startTime':startTime, 'current':currentNav, 'oneMonth':oneMonthNav, 'threeMonth':threeMonthNav, 
                              'halfYear':halfYearNav, 'oneYear':oneYearNav, 'twoYear':twoYearNav, 'bottom':bottomNav, 'top':topNav}
                             , ignore_index = True)
except BaseException as e:
    print('处理code=' + tsCode + '时出现异常', e)
    traceback.print_exc()
print(unusedCount)    
print('success')
print(funds.shape)
```

```python
from collections import Counter
import datetime


pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

startDate = funds.startTime.values
centerDay = Counter(startDate).most_common()[0][0]
centerDate = datetime.datetime.strptime(str(centerDay),'%Y%m%d')

delta = datetime.timedelta(days=-10)
dateLowBound = centerDate + delta
lowBoundDay = dateLowBound.strftime('%Y%m%d')
delta = datetime.timedelta(days=10)
dateHighBound = centerDate + delta
highBoundDay = dateHighBound.strftime('%Y%m%d')
print(centerDay, lowBoundDay, highBoundDay)

centerFunds = funds[(funds.startTime > int(lowBoundDay)) & (funds.startTime < int(highBoundDay))]

centerFunds['oneMonth'] = centerFunds.apply(lambda x:(x['current']/x['oneMonth']-1)*100, axis=1)
centerFunds['threeMonth'] = centerFunds.apply(lambda x:(x['current']/x['threeMonth']-1)*100, axis=1)
centerFunds['halfYear'] = centerFunds.apply(lambda x:(x['current']/x['halfYear']-1)*100, axis=1)
centerFunds['oneYear'] = centerFunds.apply(lambda x:(x['current']/x['oneYear']-1)*100, axis=1)
centerFunds['twoYear'] = centerFunds.apply(lambda x:(x['current']/x['twoYear']-1)*100, axis=1)
centerFunds['bottom'] = centerFunds.apply(lambda x:(x['bottom']/x['top']-1)*100, axis=1)

print(centerFunds.shape)
print(centerFunds)
```

```python
import numpy as np

print('下坡中位数', end = '') 
print(np.median(centerFunds.bottom.values))

print('下坡平均数', end = '') 
print(np.mean(centerFunds.bottom.values))

##筛选出回撤小的基金：依据2015年峰值至2016年谷底回撤控制能力 
retracement = min((np.median(centerFunds.bottom.values), np.mean(centerFunds.bottom.values))) 
print(retracement) 
stableFund = centerFunds[centerFunds.bottom > retracement] 
print(stableFund.shape)


# ##筛选短期高收益股票
# result = stableFund.sort_values(by = 'threeMonth', ascending = False).head(100)

# ##筛选长期高收益股票 
# longTermProfit = result.sort_values(by = 'twoYear', ascending = False).head(30)

##筛选中期高收益股票 
mediumTerm = longTermProfit.sort_values(by = 'oneMonth', ascending = False).head(10)


print(mediumTerm)
```
