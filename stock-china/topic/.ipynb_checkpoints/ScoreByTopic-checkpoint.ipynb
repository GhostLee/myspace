```python
import pandas as pd
import json
from pandas import DataFrame

## 本程序按题材给股票加权积分 ##
level0 = ['汽车整车', '银行', '港口航运', '化工材料', 
          '华为电视', '华为汽车', '超高清视频', '种植业', '大飞机', 
          '国产航母', '高校', '旅游', '国企改革', '文化传媒', '养老概念',
          '边缘计算', '健康中国', '工业4.0', '降税']
level1 = ['石墨烯', '海南', '摘帽', '国产软件', '高送转', '核电',
          '成渝特区', '华为海思', '无人驾驶', '物流', '农药', '人造肉', '游戏',
          '广电网', '雄安新区', '透明工厂', '新能源车', '页岩气', '光伏', '风电',
          '电力物联网', 'OLED', '创投', '一带一路', '养殖业', '军工', '贬值收益', '大数据', 
          '振兴东北', '贸易战概念', '乡村振兴', '基础建设', '石油石化', '地下管网']
level2 = ['稀土永磁', '大麻', '氢燃料电池', '芯片', '并购重组', '壳资源', '房地产', '白酒', '家用电器', '保险', '医药', 
          '云计算', '网络安全', '有色金属', '低价', '知识产权', '超跌', '操作系统', '印制电路板']
level4 = ['次新股', '创业板壳资源', '环保', '金融科技', '证券', '5G',  '黄金', '华为概念', '送转填权', '业绩增长'] 
level6 = ['区块链']

# print(len(level0))
# print(len(level1))
# print(len(level2))
# print(len(level4))
# print(len(level6))
# un = list(set(level0).union(set(level1)).union(set(level2)).union(set(level4)).union(set(level6)))
# print(len(un))

df = pd.read_csv('/data/stock_new/topics' + '.csv', header=None, names = ['code', 'name', 'topics'])

pd.set_option('display.width', 1000)  # 设置字符显示宽度
pd.set_option('display.max_rows', None)  # 设置显示最大行

result = DataFrame([], columns=['code', 'name', 'score', 'topics'])
failedCodes = []
for index, row in df.iterrows():
    try:
        if index % 200 == 0:
            print(str(index) + '个已处理...')
        score = 0
        topics = row.topics.split('[\'')[-1].split('\']')[0].split('\', \'')
        for topic in topics:
            if topic in level1:
                score = score + 1
            elif topic in level2:
                score = score + 2
            elif topic in level4:
                score = score + 4
            elif topic in level6:
                score = score + 6
        result = result.append({'code':row.code, 'name':row.iloc[1], 'score':score, 'topics':row.topics}, ignore_index = True)
    except BaseException as e:
        failedCodes.append(row.code)
        pass
    continue
final_result = result.sort_values(by="score" , ascending=False) 
print(final_result)
```

```python
filename = '区块链0628'
with pd.ExcelWriter(filename + '.xlsx') as writer:
    final_result.to_excel(writer, sheet_name='开盘啦数据')
print('success')
```
