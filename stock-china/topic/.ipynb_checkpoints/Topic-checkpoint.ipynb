```python
## 本程序通过爬取多数据源完成多题材筛选 ##
import requests
import json
import tushare as ts

## 通用函数
def getCodeWithSuffix(code):
    if code[0] == '6':
        return code + '.SH'
    else: 
        return code + '.SZ'
    
## 同花顺
def tonghuashun(tonghuashun_token, v):
    pre = 'http://www.iwencai.com/stockpick/cache?token='
    post = '&p=1&perpage='
    url = pre + tonghuashun_token + post
    headers = {'User-Agent':'Mozilla/5.0 \(Windows NT 6.1; WOW64\) AppleWebKit/537.36 \(KHTML, like Gecko\) Chrome/39.0.2171.71 Safari/537.36'}
    c = 'cid=d1ac40a4d5f4281a1116cb79aadd04d31551921949; ComputerID=d1ac40a4d5f4281a1116cb79aadd04d31551921949; ver_mark=d; PHPSESSID=e7fa610e42fe384ee1501c8ee4d525e3; guideState=1'
    cookies = {}
    aa = c.split(';')
    for line in aa:
        key, value = line.split('=', 1)
        cookies[key] = value
    cookies['v'] = v    
    response = requests.get(url + '10', headers = headers, cookies = cookies)
    totalStr = response.text.split('"total":')[-1].split(",")[0]
    final_url = url + totalStr
    response = requests.get(final_url, headers = headers, cookies = cookies)
    wccode2hq = response.text.split('"wccode2hq":')[-1].split(',"orresProcess"')[0]
    print(wccode2hq)
    wccode2hq_dict = json.loads(wccode2hq)
    tonghuashun_codes = list(wccode2hq_dict.keys())
    codes = [code.split('.')[0] for code in tonghuashun_codes]
    return codes

## 东方财富
def eastmoney(east_money_code):
    pre = 'http://nufm.dfcfw.com/EM_Finance2014NumericApplication/JS.aspx?cb=jQuery112409136126804884923_1560147446776&type=CT&token=4f1862fc3b5e77c150a2b985b12db0fd&sty=FCOIATC&js=(%7Bdata%3A%5B(x)%5D%2CrecordsFiltered%3A(tot)%7D)&cmd='
    post = '&st=(ChangePercent)&sr=-1&p=1&ps=1000'
    east_money_url = pre + east_money_code + post
    response = requests.get(east_money_url)
    result = response.text.split('{data:')[-1].split(',recordsFiltered:')[0]
    codes_str = result.split('","')
    codes = []
    for code_str in codes_str:
        code = code_str.split(',')[1].split(',')[0]
        codes.append(code)
    return codes

## 涨停温度计
def thermometer(therm_code):
    pre = 'http://q.jrjimg.cn/?q=cn|s|'
    post = '&c=m&n=hqa&o=pl,d&p='
    thermometer_url = pre + therm_code + post 
    page = 1090
    response = requests.get(thermometer_url + str(page))
    page_no = int(response.text.split('pages:', 1)[-1].split(',')[0])
    codes = []
    for i in range(0, page_no):
        final_url = thermometer_url + str(page + 1000*i)
        response = requests.get(final_url)
        codes_info_str = response.text.split('HqData:')[-1].split('};')[0]
        codes_info_list = codes_info_str.split('],\n[')
        for code_info in codes_info_list:
            code = code_info.split('","')[1].split('","')[0]
            codes.append(code) 
    return codes

## tushare
def tusharetopic(tushare_topic_id):
    token = 'c13d897473cb56b68a5d9e5476a934922478db2b3ecbe5c7efb03b65'    
    pro = ts.pro_api(token)
    df = pro.concept_detail(id='TS2', fields='ts_code,name')
    codes_withSuffix = df.ts_code.values
    codes = [code.split('.')[0] for code in codes_withSuffix]
    return codes

## 选股宝
def xuangubao(xuangubao_id):
    xuangubao_url = 'https://flash-api.xuangubao.cn/api/plate/plate_set?id=' + xuangubao_id
    response = requests.get(xuangubao_url)
    data_json = json.loads(response.text)
    stocks_json = data_json['data']['stocks']
    codes = []
    for stock_json in stocks_json:
        code = stock_json['symbol']
        codes.append(code.split('.')[0])
    return codes

## 开盘啦
def kaipanla(kaipanla_topic_id):
    kaipanla_url = 'https://pchq.kaipanla.com/w1/api/index.php'
    form_data = 'c=PCArrangeData&a=GetZSHQPlate&SelType=4,2,3,1&LType=6&LOrder=1&LStart=&LEnd=&LIndex=0&Lst=1000&UserID=570132&Token=f6b329f1fa4b35c174af28f417372a05'
    forms = {}
    form_list = form_data.split('&')
    for form in form_list:
        [key, value] = form.split('=')
        forms[key] = value;
    forms['PStockID'] = kaipanla_topic_id
    forms['PlateID'] = kaipanla_topic_id
    response = requests.post(kaipanla_url, data=forms)
    stocks_str = response.text.split('"list":[', 1)[-1].split(',"Time":')[0]
    codes = []
    stocks_str_list = stocks_str.split('],[')
    for stock_str in stocks_str_list:
        code_str = stock_str.split('"', 1)[-1].split('"', 1)[0]
        codes.append(code_str)
    return codes

## 获取密钥
def getKey(topic_name):
    if topic_name == '自主可控':
        token = '3f5d996b30f7dea8a8a12cf302e0236a'
        v = 'ArjdoL-EERSkWX2QLr99nijDj21JIRwDfoXwL_IqBPOmDVYbmjHsO86VwLNB'
        east_money_code = None
        therm_code = None
        xuangubao_id = None
        kaipanla_topic_id = '801075'
        return [token, v, east_money_code, therm_code, xuangubao_id, kaipanla_topic_id]
    elif topic_name == '集成电路':
        token = '8a204fb4df658820894586ddf534e235'
        v = 'AilMc7YzUFd9tmzjsNLcPcF8Pt6A9h2rh-pBvMsdpZBPkkdIE0Yt-Bc6UYVY'
        east_money_code = None
        therm_code = 'bk400130051'
        xuangubao_id = None
        kaipanla_topic_id = '885675'
        return [token, v, east_money_code, therm_code, xuangubao_id, kaipanla_topic_id]
    elif topic_name == '芯片':
        token = 'dada51c6b6a139cc941c0542505edf03'
        v = 'AjNWkQgNatX7ACYBWAu2fwcqxDxeaMfogfwL1-XTj9KJ5F1qbThXepHMm6T2'
        east_money_code = 'C.BK08911'
        therm_code = 'bk400129466'
        xuangubao_id = '31765737'
        kaipanla_topic_id = '885756'
        return [token, v, east_money_code, therm_code, xuangubao_id, kaipanla_topic_id]
    elif topic_name == '国产软件':
        token = '8d4ff07f05a5bc1d2b865c6d1a9c8235'
        v = 'AuOGoTidusL_onb2ZXi-jiEcdCyO2HYVsW27PRVDP1a79w2anagHasE8S5Mm'
        east_money_code = 'C.BK06961'
        therm_code = 'bk400129704'
        xuangubao_id = '16853682'
        kaipanla_topic_id = '801128'
        return [token, v, east_money_code, therm_code, xuangubao_id, kaipanla_topic_id]
    elif topic_name == '超跌':
        token = '693268268bda024c86a90ffd5052a7eb'
        v = 'Asyp7Otg_RdwE-nrgA0ZG0I5myH9BXTycqCFYyaC2AwYLmJXThVAP8K5VB91'
        east_money_code = None
        therm_code = None
        xuangubao_id = None
        kaipanla_topic_id = None
        return [token, v, east_money_code, therm_code, xuangubao_id, kaipanla_topic_id]
    elif topic_name == '次新':
        token = '01d3fdd8ec5d8a00f0ca7acbd02fdaf1'
        v = 'AkgtUE90gWNkEO3n6XbVT-5VH71Z8aw6jlWAAAL6lEO23earqgF8i95lUAdR'
        east_money_code = 'C.BK05011'
        therm_code = 'bk400122019'
        xuangubao_id = '17864537'
        kaipanla_topic_id = '801058'
        return [token, v, east_money_code, therm_code, xuangubao_id, kaipanla_topic_id]
    elif topic_name == '云计算':
        token = 'f3ff59a1a6baee385c3793d8a82bd7a2'
        v = 'AsKnzhGWK5XisDfJcX-P2ZhXFcMnk8ZoeJe6-QzY7jXgX2w19CMWvUgnCu3f'
        east_money_code = 'C.BK05791'
        therm_code = 'bk400124688'
        xuangubao_id = '17928146'
        kaipanla_topic_id = '801260'
        return [token, v, east_money_code, therm_code, xuangubao_id, kaipanla_topic_id]

def getTopicCodes(topic_name):
    [token, v, east_money_code, therm_code, xuangubao_id, kaipanla_topic_id] = getKey(topic_name)
    codes = []
    if token != None:
        aaa = tonghuashun(token, v)
        print('同花顺' + str(len(aaa)))
        codes.extend(aaa)
    if east_money_code != None:
        bbb = eastmoney(east_money_code)
        print('东财' + str(len(bbb)))
        codes.extend(bbb)
    if therm_code != None:
        ccc = thermometer(therm_code)
        print('涨停温度计' + str(len(ccc)))
        codes.extend(ccc)
    if xuangubao_id != None:
        ddd = xuangubao(xuangubao_id)
        print('选股宝' + str(len(ddd)))
        codes.extend(ddd)
    if kaipanla_topic_id != None:
        eee = kaipanla(kaipanla_topic_id)
        print('开盘啦' + str(len(eee)))
        codes.extend(eee)
    return list(set(codes))
```
